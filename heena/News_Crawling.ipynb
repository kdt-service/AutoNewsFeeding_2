# 최종코드

```python
#여러 날짜 여러 페이지(최종) - 네이버 사회의 사건사고 카테고리

a=1
dates = list(range(20230201, 20230229)) + list(range(20230301, 20230317)) #크롤링하고자 하는 날짜
news_list = []
for date in dates:
    URL = "https://news.naver.com/main/list.naver?mode=LS2D&sid2=249&sid1=102&mid=shm&date={}".format(date)
    
    
    
    #총 페이지 수 구하기
    URL = URL + "&page={}"
    page = 1

    while True:
        response = requests.get(URL.format(page), headers=headers)    
        soup = BeautifulSoup(response.text, 'html.parser')
        try: 
            if soup.select_one('.next').text == '다음':
                page += 1
                print(f'완료: {page}')
        except:
            break

    for_pages = soup.find(class_="paging" )
    pages = for_pages.find_all('a')
    plus_page = len(pages) - 1
    
    print(f'마지막 페이지 화면: {plus_page}')
    
    total_page = page + plus_page
    
    print(f'총 페이지는? {total_page} !!!!! ')

    
    
    for page in range(1,total_page+1):

        response = requests.get(URL.format(page), headers=headers)    
        soup = BeautifulSoup(response.text, 'html.parser')

        url_in = soup.select('div[id="main_content"] li') #한 페이지 최대 20개
        url_list = []
        for li in url_in:
            url_one = li.select_one('dt a[href]').attrs['href'] #기자와 작성시간 알기 위해 기사 링크 추가
            url_list.append(url_one)
            

        print(f'{date} {page}페이지 링크 다 가져옴!!')    


        
        for url in url_list: #한 링크씩 들어가기
            url = url
            response = requests.get(url, headers=headers)
            soup = BeautifulSoup(response.text, 'html.parser')
            news_dict = {}
            news_dict['platform'] = 'Naver'
            news_dict['main_category'] = '사회'
            news_dict['sub_category'] = '사건사고'
            news_dict['title'] = soup.select_one('.media_end_head_headline').text.strip()
            #news_dict['press'] = soup.select_one('.media_end_head_title span').text.strip()
            news_dict['content'] = soup.select_one('div[id="newsct_article"]').text.strip()
            try:
                news_dict['writer'] = soup.select_one('.media_end_head_journalist_name').text.strip()
            except:
                try:
                    news_dict['writer'] = soup.select_one('.byline_s').text[:6]
                except:
                    news_dict['writer'] = None
            #time = soup.select_one('.media_end_head_info_datestamp_time').text.strip()
            #news_dict['writed_at'] = datetime.datetime.strptime(time, '%Y-%m-%d %H:%M:%S')
            news_dict['writed_at'] = soup.select_one('.media_end_head_info_datestamp_time').text.strip()
            news_list.append(news_dict)
            
            print('dict 하나 추가함!!!')
            
            
        print(f'{page} dict 추가 완~!~!')
        
        
    print(f'{date} 데이터 추가 완료~~~~~!!!!!!!!!')        
    a+=1        
    if a == 3:
        break
```
