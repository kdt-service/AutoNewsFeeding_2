```python
import pandas as pd
```

```python
# 희나
social1 = pd.read_csv('../data/news_social.csv', index_col=0)
```

```python
# 정현
df1 = pd.read_csv('../data/사회_식품의료.csv', encoding='utf-8-sig', usecols=list(range(0,8)))
df2 = pd.read_csv('../data/사회_지역.csv', encoding='utf-8-sig', usecols=list(range(0,8)))
```

```python
# 연주
df3 = pd.read_csv('../data/사회_언론.csv', encoding='utf-8-sig', usecols=list(range(0,8)))
df4 = pd.read_csv('../data/사회_인권복지.csv', header=None, encoding='utf-8-sig', usecols=list(range(0,8)), names=list(df3.columns))
```

```python
# 병합
social1 = social1[['title', 'writer', 'writed_at', 'content', 'main_category', 'sub_category', 'platform']]

df = pd.concat([df1, df2], axis=0)
df = df[['title', 'writer', 'writed_at', 'content', 'main_category', 'sub_category', 'platform']]

new_df = pd.concat([df3, df4], axis=0)
new_df = new_df[['title', 'writer', 'writed_at', 'content', 'main_category', 'sub_category', 'platform']]

social_df = pd.concat([social1, df, new_df], axis=0)
social_df
```

```python
# 저장
social_df.to_csv('../data/naver_social.csv', encoding='utf-8-sig', index=False)
```

```python
social_df = pd.read_csv('../data/naver_social.csv', encoding='utf-8-sig')
```

```python
# 데이터 처리
social_df[social_df['writed_at'].str.contains('최영')]
social_df.drop(index=24912, axis=0, inplace=True)
```

```python
# 날짜 수정
for i in social_df[social_df['writed_at'].str.contains('기사입력')].index:
    social_df.loc[i, 'writed_at'] = ' '.join(social_df.loc[i, 'writed_at'].split('\n')[0].split(' ')[1:])
```

```python
social_df['writed_at'] = social_df['writed_at'].str.replace('오후', 'PM')
social_df['writed_at'] = social_df['writed_at'].str.replace('오전', 'AM')
social_df['writed_at'] = pd.to_datetime(social_df['writed_at'], format='%Y.%m.%d. %p %I:%M')
social_df['writed_at'] = social_df['writed_at'].apply(lambda x : x.strftime('%Y-%m-%d %H:%M:%S'))

social_df = social_df.sort_values(by=['sub_category', 'writed_at']).reset_index(drop=True)
```

```python
# 플랫폼 이름 수정
social_df['platform'] = '네이버'
```

```python
social_df.to_csv('../data/naver_social.csv', encoding='utf-8-sig', index=False)
```
